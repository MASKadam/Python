```python
#Q14 Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
# Perform the following tasks using Matplotlib

# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
# right y-axis for 'Humidity').

# b) Label the x-axis as 'Date'.

# c) Set the title of the plot as 'Temperature and Humidity Over Time'.



# Create a time-series dataset
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')
temperature = np.random.uniform(low=20, high=35, size=len(date_range))
humidity = np.random.uniform(low=30, high=90, size=len(date_range))

data = {
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
}

df = pd.DataFrame(data)

# Plot Temperature and Humidity on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create another y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Display the legend
fig.tight_layout()  # Adjust layout to make room for the labels and title
fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))

# Show the plot
plt.show()
```

```python
#Q15 Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
# tasks using Matplotlib:

# a) Plot a histogram of the data with 30 bins.

# b) Overlay a line plot representing the normal distribution's probability density function (PDF).

# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

# d) Set the title of the plot as 'Histogram with PDF Overlay'.

from scipy.stats import norm
#Ans - 
# Create a NumPy array containing 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Overlay a line plot representing the normal distribution's probability density function (PDF)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2, label='Normal PDF')

# Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title of the plot as 'Histogram with PDF Overlay'
plt.title('Histogram with PDF Overlay')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
#Q17  Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
#      origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.


# Create two random arrays
x = np.random.uniform(-10, 10, 100)
y = np.random.uniform(-10, 10, 100)

# Create a DataFrame with quadrant information
df = pd.DataFrame({
    'X': x,
    'Y': y,
    'Quadrant': np.where((x >= 0) & (y >= 0), 'Quadrant 1',
                np.where((x < 0) & (y >= 0), 'Quadrant 2',
                np.where((x < 0) & (y < 0), 'Quadrant 3', 'Quadrant 4')))
})

# Create a scatter plot using Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='viridis')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Display the legend
plt.legend(title='Quadrants')

# Show the plot
plt.show()
```

```python
#Q18  With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import Range1d
import numpy as np

# Display the plot inline in the notebook
output_notebook()

# Generate data for the sine wave
x = np.linspace(0, 4 * np.pi, 1000)
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)')

# Add a line renderer with the sine wave data
p.line(x, y, line_width=2)

# Add grid lines
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'

# Set the ranges for x and y axes
p.x_range = Range1d(0, 4 * np.pi)
p.y_range = Range1d(-1.5, 1.5)

# Show the results
show(p)
```

```python
#19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, 
#     add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart.

#Ans:

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Viridis5
import numpy as np
import pandas as pd

# Use output_notebook to display the plot inline in Jupyter notebooks
output_notebook()

# Generate random categorical data 
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 101, size=len(categories))

# Generate a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Create a new plot with a title and axis labels
p = figure(x_range=categories, title="Random Categorical Bar Chart", 
           x_axis_label='Categories', y_axis_label='Values', 
           plot_height=400, plot_width=600, tools='hover')

# Add a bar renderer
p.vbar(x='categories', top='values', width=0.9, source=source, 
       line_color='white',
       fill_color=factor_cmap('categories', palette=Viridis5, factors=categories))

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Customize legend properties
p.legend.title = "Categories"
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the plot
show(p)
```

```python
#20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.

import plotly.graph_objects as go
import numpy as np

np.random.seed(42) 
x_values = np.arange(10)  
y_values = np.random.randint(1, 101, size=len(x_values)) 

# Create a line plot using Plotly
fig = go.Figure(data=go.Scatter(x=x_values, y=y_values, mode='lines+markers'))

# Customize the layout
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X Axis Label',
    yaxis_title='Y Axis Label'
)

fig.show()
```

```python
#21. Using Plotly, 
#   create an interactive pie chart of randomly generated data, add labels and percentages, set the title as 'Interactive Pie Chart'.

#Ans

import plotly.graph_objects as go
import numpy as np


np.random.seed(42)  
labels = ['A', 'B', 'C', 'D', 'E'] 
values = np.random.randint(1, 101, size=len(labels))  

# Create a pie chart using Plotly
fig = go.Figure(data=[go.Pie(
    labels=labels,
    values=values,
    hole=0.3,  
    textinfo='label+percent',  
    pull=[0.1, 0, 0, 0, 0], 
    marker=dict(colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']) 
)])


fig.update_layout(
    title='Interactive Pie Chart',
    template='plotly_white' 
)

fig.show()

```
